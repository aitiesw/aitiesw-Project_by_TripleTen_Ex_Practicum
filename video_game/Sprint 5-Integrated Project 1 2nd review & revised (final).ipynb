Which plantforms are growing or shrinking?

```python
# Pivot the data to have years as columns and platforms as rows, and total sales as values
platform_year_sales = df.groupby(['platform', 'year_of_release'])['global_sales'].sum().reset_index()
pivot_table = platform_year_sales.pivot_table(index='platform', columns='year_of_release', values='global_sales', fill_value=0)

# Calculate the change in sales for each platform over consecutive years
sales_change = pivot_table.diff(axis=1)

# Identify growing and shrinking platforms
growing_platforms = sales_change[sales_change > 0].count(axis=1)
shrinking_platforms = sales_change[sales_change < 0].count(axis=1)

# Sort the growing and shrinking platforms based on their counts
sorted_growing_platforms = growing_platforms.sort_values(ascending=False)
sorted_shrinking_platforms = shrinking_platforms.sort_values(ascending=False)

display("Sorted Growing Platforms:", sorted_growing_platforms)
display("\nSorted Shrinking Platforms:", sorted_shrinking_platforms)
```

We identify growing platforms by counting the number of positive changes in sales over consecutive years for each platform. Similarly, we identify shrinking platforms by counting the number of negative changes in sales.

This will give us a count of how many times each platform's sales have grown or shrunk over consecutive years. Platforms with higher counts of positive changes are growing, while platforms with higher counts of negative changes are shrinking.

PC, GB, PS3 & NES are growing platforms while PC, DS, PSP, PSP2 are the shrinking platforms. PC appears as growing and shrinking platform. It might due to the sales of some games are growing fast and while other games are slow selling.

## Hypotheses Testing

A two sample t-test is used to test whether or not the means of two populations are equal.

This type of test makes the following assumptions about the data:

1. Independence: The observations in one sample are independent of the observations in the other sample.

2. Normality: Both samples are approximately normally distributed. If the sample sizes are small (n < 50), then we can use a Shapiro-Wilk test to determine if each sample size is normally distributed. If the p-value of the test is less than a certain significance level, then the data is likely not normally distributed.
If the sample sizes are large, then it’s better to use a Q-Q plot to visually check if the data is normally distributed.

3. Homogeneity of Variances: Both samples have approximately the same variance. We use the following rule of thumb to determine if the variances between the two samples are equal: If the ratio of the larger variance to the smaller variance is less than 4, then we can assume the variances are approximately equal and use the two sample t-test.

4. Random Sampling: Both samples were obtained using a random sampling method.

If one or more of these assumptions are violated, then the results of the two sample t-test may be unreliable or even misleading.

```python
# Filter the data for Xbox One and PC platforms
xbox_one_data = refined_df[refined_df['platform'] == 'XOne']
pc_data = refined_df[refined_df['platform'] == 'PC']
```

```python
# Create histograms for both datasets to check for normality
plt.hist(pc_data['user_score'], bins=30, alpha=0.5, label='PC')
plt.hist(xbox_one_data['user_score'], bins=30, alpha=0.5, label='Xbox One')

# Add labels and a legend
plt.xlabel('User Score')
plt.ylabel('Frequency')
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

From the above plotting, we can see the data is left-skewed for both Xbox One & PC. 

As the sample sizes are large (500 for our study), we will use a Q-Q plot to visually check if the data is normally distributed.

From the Q-Q plot, he data points roughly fall along a straight diagonal line in a Q-Q plot, then the datasets likely follow a normal distribution.

### Average user ratings of the Xbox One and PC platforms are the same

Formulate Hypotheses:

Null Hypothesis (H0): The average user ratings of the Xbox One and PC platforms are the same.

Alternative Hypothesis (H1): The average user ratings of the Xbox One and PC platforms are different.

```python
# Perform a two-sample t-test
t_statistic, p_value = stats.ttest_ind(xbox_one_data['user_score'].dropna(), pc_data['user_score'].dropna())

# Set the significance level (alpha). Alpha value of 0.05 means that we are willing to accept a 5% chance of making a Type I error, which is the probability of rejecting a null hypothesis when it is actually true. In other words, we are comfortable with a 5% chance of concluding there is an effect when there is none.
alpha = 0.05

# Print the results
display("T-statistic:", t_statistic)
display("P-value:", p_value)

# Check if the p-value is less than alpha to determine statistical significance
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in user ratings between Xbox One and PC.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in user ratings between Xbox One and PC.")
```
