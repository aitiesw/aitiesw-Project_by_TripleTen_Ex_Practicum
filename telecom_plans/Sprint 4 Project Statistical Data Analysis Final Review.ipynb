# Which one is a better plan?

Telecom operator Megaline offers its clients two prepaid plans, Surf and Ultimate. The commercial department wants to know which of the plans brings in more revenue in order to adjust the advertising budget.

We are going to carry out a preliminary analysis of the plans based on a relatively small client selection. We have the data on 500 Megaline clients: who the clients are, where they're from, which plan they use, and the number of calls they made and text messages they sent in 2018. Our job is to analyze the clients' behavior and determine which prepaid plan brings in more revenue.

## Aggregate data per user

```python
# Calculate the number of calls made by each user per month. Save the result.
grouped_calls_per_month_by_user = df_calls.groupby(['user_id', 'month'])['id'].count().reset_index()

# Rename the 'id' column to 'call_count'
grouped_calls_per_month_by_user.rename(columns={'id': 'total_call_count'}, inplace=True)

grouped_calls_per_month_by_user
```

```python
# Calculate the amount of minutes spent by each user per month. Save the result.
grouped_minutes_per_month_by_user = df_calls.groupby(['user_id', 'month'])['duration_rounded_up'].sum().reset_index()

# Rename the 'id' column to 'call_count'
grouped_minutes_per_month_by_user.rename(columns={'duration_rounded_up': 'total_call_time_min'}, inplace=True)

grouped_minutes_per_month_by_user
```

```python
# Calculate the number of messages sent by each user per month. Save the result.
grouped_message_per_month_by_user = df_messages.groupby(['user_id', 'month'])['id'].count().reset_index()

# Rename the 'id' column to 'call_count'
grouped_message_per_month_by_user.rename(columns={'id': 'total_messages'}, inplace=True)

grouped_message_per_month_by_user
```

```python
# Calculate the volume of internet traffic used by each user per month. Save the result.
grouped_internet_per_month_by_user = df_internet.groupby(['user_id', 'month'])['mb_used'].sum().reset_index()

# Rename the 'id' column to 'call_count'
grouped_internet_per_month_by_user.rename(columns={'mb_used': 'total_internet_traffic_mb'}, inplace=True)

grouped_internet_per_month_by_user
```

Calculating monthly revenue for each user:

```python
# Calculate the monthly revenue for each user
def calculate_monthly_revenue(row):
    calls_excess = max(row['total_call_time_min'] - row['minutes_included'], 0)
    messages_excess = max(row['total_messages'] - row['messages_included'], 0)
    internet_excess_gb = max(row['total_internet_traffic_gb'] - row['gb_per_month_included'], 0)
    
    total_revenue = (
        row['usd_monthly_pay'] + 
        calls_excess * row['usd_per_minute'] + 
        messages_excess * row['usd_per_message'] + 
        internet_excess_gb * row['usd_per_gb']
    )
    
    return total_revenue

final_merged_data_with_plans['monthly_revenue'] = final_merged_data_with_plans.apply(calculate_monthly_revenue, axis=1)

display(final_merged_data_with_plans[['user_id', 'month', 'plan_name', 'usd_monthly_pay', 'monthly_revenue']])
```

```python
# Counting monthly revenue.
monthly_revenue = final_merged_data_with_plans.groupby(['month', 'plan_name'])['monthly_revenue'].sum().reset_index()
monthly_revenue
```

## Study user behaviour

We calculated some useful descriptive statistics for the aggregated and merged data, which typically reveal an overall picture captured by the data. Our objective is to compare the plans and decide on which one is more profitable.

### Calls

```python
# Group data by 'plan_name' and 'month' and calculate the average call duration
average_call_duration_by_plan_month = final_merged_data_with_plans.groupby(['plan_name', 'month'])['total_call_time_min'].mean().reset_index()

# Pivot the data for plotting
pivot_data_average_call_duration = average_call_duration_by_plan_month.pivot(index='month', columns='plan_name', values='total_call_time_min')

# Plotting
pivot_data_average_call_duration.plot(kind='bar', figsize=(14, 6))
plt.xlabel('Month')
plt.ylabel('Average Call Duration')
plt.title('Average Call Duration per Plan per Month')
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

The mean call duration for ultimate users are lower than surf users in January, March, April, June, July & December.

```python
# Compare the number of minutes users of each plan require each month. Plot a histogram.
total_minutes_by_plan_month = final_merged_data_with_plans.groupby(['plan_name', 'month'])['total_call_time_min'].sum().reset_index()

# Plot histograms for each plan's monthly minutes
for plan in final_merged_data_with_plans['plan_name'].unique():
    plan_data = final_merged_data_with_plans[final_merged_data_with_plans['plan_name'] == plan]['total_call_time_min']
    plt.hist(plan_data, bins=50, alpha=0.8, label=plan)

plt.title('Distribution of Monthly Call Time by Plan')
plt.xlabel('Monthly Call Time')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

The total monthly call of surf plan is peaked around 500 minutes while ultimate plate is peaked around 400 minutes. Both plans are left skewed, which mean less users are calling longer time in total.

## Test statistical hypotheses

To test the hypothesis that the average revenue from users of the Ultimate and Surf calling plans differs, we perform a two-sample t-test for independent samples. 

Hypotheses:

Null Hypothesis (H0): The average revenue from users of the Ultimate and Surf calling plans is the same.
Alternative Hypothesis (H1): The average revenue from users of the Ultimate and Surf calling plans differs.
Statistical Test:

A two-sample t-test for independent samples is appropriate for comparing the means of two groups.

Alpha Value:

The alpha value, also known as the significance level, determines the threshold for rejecting the null hypothesis. A common alpha value is 0.05 (5%), which means we are willing to accept a 5% chance of making a Type I error (rejecting a true null hypothesis).

```python
# Test the hypotheses

# Select revenue data for Ultimate and Surf plans
revenue_ultimate = final_merged_data_with_plans[final_merged_data_with_plans['plan_name'] == 'ultimate']['monthly_revenue']
revenue_surf = final_merged_data_with_plans[final_merged_data_with_plans['plan_name'] == 'surf']['monthly_revenue']

# Perform two-sample t-test
t_statistic, p_value = ttest_ind(revenue_ultimate, revenue_surf)

alpha = 0.05  
# alpha value 0.05 (5%), which means we're willing to accept a 5% chance of making a Type I error (rejecting a true null hypothesis).


print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

if p_value < alpha:
    print("Reject Null Hypothesis: The revenue from ultimate and surf plans differs.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference in revenue.")
```

1. The positive T-statistic (5.57) indicates that the average revenue of the Ultimate plan is higher than the Surf plan.
2. The results of the t-test indicate a very low p-value (2.86e-08), which is significantly smaller than the typical alpha value of 0.05. This suggests that we should reject the null hypothesis. In other words, the average revenue from users of the ultimate and surf calling plans is statistically significantly different.

To test the hypothesis that the average revenue from users in the NY-NJ area is different from that of the users from the other regions, we perform a two-sample t-test for independent samples. 

Hypotheses:

Null Hypothesis (H0): The average revenue from users in the NY-NJ area is the same as that of users from other regions.
Alternative Hypothesis (H1): The average revenue from users in the NY-NJ area is different from that of users from other regions.
Statistical Test:

A two-sample t-test for independent samples is appropriate for comparing the means of two groups.

Alpha Value:

The alpha value, also known as the significance level, determines the threshold for rejecting the null hypothesis. A common alpha value is 0.05 (5%), which means we are willing to accept a 5% chance of making a Type I error (rejecting a true null hypothesis).

```python
# Test the hypotheses

# Select revenue data for NY-NJ and Other regions
revenue_nynj = final_merged_data_with_plans[final_merged_data_with_plans['region'] == 'NY-NJ-PA']['monthly_revenue']
revenue_other = final_merged_data_with_plans[final_merged_data_with_plans['region'] != 'NY-NJ-PA']['monthly_revenue']

# Perform two-sample t-test
t_statistic, p_value = ttest_ind(revenue_nynj, revenue_other)

alpha = 0.05  # chosen alpha value

print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

if p_value < alpha:
    print("Reject Null Hypothesis: The average revenue from NY-NJ area users is different from other regions.")
else:
    print("Fail to Reject Null Hypothesis: No significant difference in average revenue between NY-NJ area and other regions.")
```

The results of the t-test indicate p-value of 0.04, which is smaller than the typical alpha value of 0.05. This suggests that we should reject the null hypothesis. In other words, the average revenue from users in the NY-NJ area is different from that of the users from the other regions.
