### Check if Recovery is Calculated Correctly

We use the training set to calculate recovery for the rougher.output.recovery feature and calculated the MAE between the calculations and the feature values.

To calculate recovery, we'll use the following formula:

Recovery= C×(F−T) / F×(C−T) × 100%

Where:
C is the concentration of the valuable metal in the concentrate (rougher.output.concentrate_au).
F is the concentration of the valuable metal in the feed (rougher.input.feed_au).
T is the concentration of the valuable metal in the tailings (rougher.output.tail_au).

```python
# Create function to calculate Recovery.
def recovery(df, C, F, T):
    rec = ((df[C] * (df[F] - df[T])) / (df[F] * (df[C] - df[T]))) * 100
    return rec

# Carry out calculations for rows which don't have missing values in rougher.output.recovery.
recovery_pred = recovery(df_train.dropna(subset=['rougher.output.recovery']), 'rougher.output.concentrate_au', 'rougher.input.feed_au', 'rougher.output.tail_au')
recovery_act = df_train['rougher.output.recovery'].dropna().to_list()



# Perform paired sample t-test
t_statistic, p_value = stats.ttest_rel(recovery_pred, recovery_act)

# Print the t-statistic and p-value
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Set significance level (alpha)
alpha = 0.05

# Compare p-value to significance level
if p_value < alpha:
    print("The difference between recovery_pred and recovery_act is statistically significant.")
else:
    print("The difference between recovery_pred and recovery_act is not statistically significant.")
```

Once we calculated the recovery values, we use Paired Sample T-test to compare if the recover_pred is different from recovery_act.

In conclusion, the difference between recovery_pred and recovery_act is not statistically significant.

### Total concentrations comparison of all substances at different stages: raw feed, rougher concentrate, and final concentrate.

```python
# Create function to get total concentration of all metals at certain stages

def metal_conc_sum(df, stage):
    metal_handles = list(metals.keys())
    total_conc = df[stage+metal_handles[0]] + df[stage+metal_handles[1]] + df[stage+metal_handles[2]]
    return total_conc
```

```python
# Create plot to visualise total concentration of all metals at certain stages

fig = plt.figure(figsize=(13, 6))
fig.suptitle('Total Concentration of All Metals by Stage (Density)', fontsize=15, y=1)

sns.kdeplot(metal_conc_sum(df_full, 'rougher.input.feed'), label='rougher.input.feed', fill=True)
sns.kdeplot(metal_conc_sum(df_full, 'rougher.output.concentrate'), label='rougher.output.concentrate', fill=True)
sns.kdeplot(metal_conc_sum(df_full, 'final.output.concentrate'), label='final.output.concentrate', fill=True)
plt.legend(fontsize=10)
plt.xlabel('Total Metal Concentration (%)', fontsize=12, labelpad = 15)
plt.ylabel('Density', fontsize=12, labelpad=15)
    
plt.tight_layout()
plt.show()
```

Above plot shows the density for the distribution of total metal concentrations for each stage of the purification process. As we can see, the total concentration for all metals increases, for the most part, at each stage of the purification process. However, there are clear outliers around 0. In order to ensure that these outliers do not negatively affect the models we will train in a later section, we will remove the outliers from our training and testing datasets. Based on the plot, a good cutoff concentration value is 10 for each stage. We will filter these outliers from the df_full DataFrame, then remove these rows from the other datasets using the 'date' indexer.

## Model Building

We will create a function to calculate the sMAPE (Symmetric Mean Absolute Percentage Error) metric. We will then train and test models with the goal of predicting the amount of gold recovered from gold ore.

### Write a function to calculate the final sMAPE (Symmetric Mean Absolute Percentage Error) value.

```python
# Create function to calculate sMAPE
def smape2(y_true, y_pred):
    smape = np.mean(abs(y_pred - y_true) / ((abs(y_true) + abs(y_pred)) / 2)) * 100 
    return smape

# Create function to calculate final sMAPE
def final_smape(y_true, y_pred):
    predicted_rough, predicted_final = y_pred[:, 0], y_pred[:, 1]
    true_rough, true_final = y_true[:, 0], y_true[:, 1]
    final_smape = (0.25 * (smape2(true_rough, predicted_rough))) + (0.75 * (smape2(true_final, predicted_final)))
    return final_smape
```

### Separate Features & Targets for Training & Testing Datasets

```python
# Get the lists of feature and training columns
targets = ['rougher.output.recovery', 'final.output.recovery']
features = [x for x in df_train if x not in targets and x != 'date']

# Split training and testing datasets by features and targets
y_train, y_test = df_train[targets].values, df_test[targets].values
X_train, X_test = df_train[features].values, df_test[features].values
```

### Train different models. Evaluate them using cross-validation. Pick the best model and test it using the test sample. Provide findings.

#### Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor

# Create a loop for max_depth from 1 to 10
for depth in range(1, 11):
    model = DecisionTreeRegressor(random_state=12345, max_depth=depth)  # Create a model, specify max_depth=depth

    # Train the model
    model.fit(X_train, y_train)
    predictions_valid = model.predict(X_test)  # Find the predictions using test set

    print("max_depth =", depth, ": ", end='')
    print(final_smape(y_test, predictions_valid))
```

1. Lower final_smape values are better. They indicate a better fit between the predicted and actual values.

2. Among the different max_depth values, the max_depth = 1 gives the lowest final_smape, which means that it's providing the best fit to the data.

3. As max_depth increases above 1, the final_smape starts to increase. This suggests that deeper trees might be overfitting to the training data.

#### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

# Create a Random Forest Regressor
model = RandomForestRegressor(random_state=12345, n_estimators=100, max_depth=1)  

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
predictions_valid = model.predict(X_test)

# Calculate f_smape
result = final_smape(y_test, predictions_valid)
print("final_smape:", result)
```

The lowest estimators numbers that gave us lowest final_smape is 100. We will use max_depth 1 as concluded from previous decision tree regressor evaluation.

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression

# Create a Linear Regressor
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
predictions_valid = model.predict(X_test)

# Calculate f_smape
result = final_smape(y_test, predictions_valid)
print("final_smape:", result)
```

We trained 3 different models and obtained the final_smape scores using the predictions of each model. The results showed that the Random Forest Regressor performed the best, with an average symmetric mean absolute percentage error of approximately 6.92. We will next to cross validate these results by using sklearn's 'cross_val_score' function.

### Models Cross-Validation

```python
from sklearn.metrics import make_scorer

# Create smape scorer
smape = make_scorer(final_smape, greater_is_better=False)

# Define the models
models = [
    DecisionTreeRegressor(random_state=12345, max_depth=1),
    RandomForestRegressor(random_state=12345, n_estimators=100, max_depth=1),
    LinearRegression()
]

# Define the names of the models for printing
model_names = ['Decision Tree', 'Random Forest', 'Linear Regression']

# Iterate through models
for i, model in enumerate(models):
    scores = abs(cross_val_score(model, X_train, y_train, cv=5, scoring=smape))
    print(f"{model_names[i]} final_smape scores:", scores)
    print(f"Mean final_smape for {model_names[i]}:", scores.mean())
    print()
```

From the above calculation, the results we obtained from cross validation are the same with the scores we received from the models we trained. Random Forest Regressor has the lowest final_smape. In obtaining these scores, we used sklearn's cross_val_score function and ran 5-fold cross validation on all of the previously used estimators. Then we passed each estimator the parameters that we found to calculate the best sMAPE score. 

In the next subsection, we will use the Random Forest Regressor model that provided us with the best sMAPE score on our testing datasets.

### Final Model Selection
