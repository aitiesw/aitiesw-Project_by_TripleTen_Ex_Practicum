## Checking the Hypotheses


**Is there a correlation between having children and paying back on time?**

```python
# Check the children data and paying back on time

correlation_children = df['children'].corr(df['debt'])
print(correlation_children)
```

```python
# Create a groupby object (i.e. children group) and aggregate the select data (i.e. debt) using the aggregation function mean.
children_default_rate = (
    df
    .groupby("children")
    .agg({"debt": "mean"})
    .reset_index()
    .rename(columns={"debt": "default_rate"})
)
# Calculate the percentage default
children_default_rate["percent_default"] = children_default_rate["default_rate"] * 100
display(children_default_rate)

def get_default_rate(dataframe, column):
    # Create a groupby object (i.e. children group) and aggregate the select data (i.e. debt) using the aggregation function mean.
    data = (
        dataframe
        .groupby(column)
        .agg({"debt": "mean"})
        .reset_index()
        .rename(columns={"debt": "default_rate"})
    )
    # Calculate the percentage default
    data["percent_default"] = (data["default_rate"] * 100).round(2)
    return data

# Example of the function call

get_default_rate(df, "children")
```

**Is there a correlation between family status and paying back on time?**

```python
# Check the family status data and paying back on time



# Calculating default-rate based on family status

# Group the filtered data by 'family_status' and calculate the default rate for each group
default_rate_by_family_status_mean = df.groupby('family_status')['debt'].mean().reset_index()
default_rate_by_family_status_median = df.groupby('family_status')['debt'].median().reset_index()

# Calculate the percentage of default 
default_rate_by_family_status_mean['Default_Rate_Percentage'] = default_rate_by_family_status_mean['debt'] * 100
default_rate_by_family_status_median['Default_Rate_Percentage'] = default_rate_by_family_status_median['debt'] * 100

# Sort the data by default rate in descending order
default_rate_by_family_status_mean.sort_values(by='Default_Rate_Percentage', ascending=False, inplace=True)
default_rate_by_family_status_median.sort_values(by='Default_Rate_Percentage', ascending=False, inplace=True)


# Display the default rates based on the number of children with debt
print(default_rate_by_family_status_mean)
print('\n')
print(default_rate_by_family_status_median)
```

**Conclusion**

The highest default rate category is unmarried and the least is widow/widower. It seems like no significant correlation between family status and paying back on time.

**Is there a correlation between income level and paying back on time?**

```python
# Check the income level data and paying back on time



# Calculating default-rate based on income level

# Calculate the default rate
default_rate_by_income_mean = df.groupby('total_income_category')['debt'].mean().reset_index()
default_rate_by_income_median = df.groupby('total_income_category')['debt'].median().reset_index()

# Calculate the percentage of default 
default_rate_by_income_mean['Default_Rate_Percentage'] = default_rate_by_income_mean['debt'] * 100
default_rate_by_income_median['Default_Rate_Percentage'] = default_rate_by_income_median['debt'] * 100

# Sort the data by default rate in descending order
default_rate_by_income_mean.sort_values(by='Default_Rate_Percentage', ascending=False, inplace=True)
default_rate_by_income_median.sort_values(by='Default_Rate_Percentage', ascending=False, inplace=True)


# Display the default rates based on the number of children with debt
print(default_rate_by_income_mean)
print('\n')
print(default_rate_by_income_median)
```
